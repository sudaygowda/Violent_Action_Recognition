### Imports and setup

```python
import pandas as pd
import sklearn.model_selection import train_test_split
from pathlib import Path
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, Dataset
from transformers import CLIPProcessor, CLIPModel
import decord
from decord import VideoReader, cpu
import numpy as np
from tqdm import tqdm
import os
```

### Choose and Load Pre-Trained VLM

```python
model_name = "openai/clip-vit-base-patch32"
processor = CLIPProcessor.from_pretrained(model_name)
base_model = CLIPModel.from_pretrained(model_name)
```

### Checking for GPU
(CUDA device will be chosen if available or else CPU will be chosen)

```python
#Checking if GPU is available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
base_model.to(device)
print(f"Using device: {device}")
```

### Load Dataset from Directories and CSVs

Defining the root path of the data

```python
#Root directory of the dataset
data_root = Path("/raw_data") 
```

Function to collect video paths with binary labels

```python
def collect_videos_with_labels(root_dir):
    video_paths = []
    labels = []

    #Violent videos(Label 1)
    violent_dir = root_dir / "Violent"
    for cam_dir in ["cam1", "cam2"]:
        cam_path = violent_dir/cam_dir
        if cam_path.exists():
            for file in cam_path.glob(".mp4"):
                video_paths.append(str(file))
                labels.append(1)
    
    #Non-Violent videos(Label 0)
    non_violent_dir = root_dir / "NonViolent"
    for cam_dir in ["cam1", "cam2"]:
        cam_path = non_violent_dir/cam_dir
        if cam_path.exists():
            for file in cam_path.glob(".mp4"):
                video_paths.append(str(file))
                labels.append(0)

    return video_paths, labels
```

Collecting all videos and labels

```python
all_video_paths, all_labels = collect_videos_with_labels(data_root)
```

Loading CSVs for additional metadata (action classes)

```python
violent_csv_path = data_root / "violent-action-classes.csv"
nonviolent_csv_path = data_root / "nonviolent-action-classes.csv"
occurrences_csv_path = data_root / "action-class-occurrences.csv"
```

Loading if they exist

```python
action_metadata = {}
if violent_csv_path.exists():
    violent_df = pd.read_csv(violent_csv_path, sep=";")
    for _, row in violent_df.iterrows():
        action_metadata[row["FILE"]] = {"classes": row["ACTION CLASSES"], "violent": 1}

if nonviolent_csv_path.exists():
    nonviolent_df = pd.read_csv(nonviolent_csv_path, sep=";")
    for _, row in nonviolent_df.iterrows():
        action_metadata[row["FILE"]] = {"classes": row["ACTION CLASSES"], "violent": 0}

if occurrences_csv_path.exists():
    occurrences_df = pd.read_csv(occurrences_csv_path, sep=";")
    print("Action Class Occurrences:\n", occurrences_df)
```

Split into train and validation (80:20)

```python
train_videos, val_videos, train_labels, val_labels = train_test_split(
    all_video_paths, all_labels, test_size=0.2, stratify=all_labels, random_state=42
```

print stats

```python
print(f"Total videos: {len(all_video_paths)} (Violent: {sum(all_labels)}, Non-Violent: {len(all_labels) - sum(all_labels)})")
print(f"Train set: {len(train_videos)} | Val set: {len(val_videos)}")
```

### Define Custom Dataset

```python
class ActionDataset(Dataset):
    def __init__(self, video_paths, labels, processor, num_frames=8):
        self.video_paths = video_paths  # List of full video file paths
        self.labels = labels  # List of 0 or 1
        self.processor = processor
        self.num_frames = num_frames

    def __len__(self):
        return len(self.video_paths)

    def __getitem__(self, idx):
        video_path = self.video_paths[idx]
        vr = VideoReader(video_path, ctx=cpu(0))
        total_frames = len(vr)
        frame_indices = np.linspace(0, total_frames - 1, self.num_frames, dtype=int)
        frames = vr.get_batch(frame_indices).asnumpy()  # Shape: (num_frames, H, W, C)
        
        # Process frames with CLIP processor
        inputs = self.processor(images=list(frames), return_tensors="pt", padding=True)
        label = torch.tensor(self.labels[idx], dtype=torch.float)
        
        return inputs, label

# Create datasets
train_dataset = ActionDataset(train_videos, train_labels, processor)
val_dataset = ActionDataset(val_videos, val_labels, processor)
```

### Define Fine-Tuned Model

```python
class FineTunedVLM(nn.Module):
    def __init__(self, base_model):
        super().__init__()
        self.vision_model = base_model.vision_model
        self.projection_dim = base_model.config.projection_dim
        self.classifier = nn.Linear(self.projection_dim, 1)  # Binary output

    def forward(self, pixel_values):
        # pixel_values shape: (batch_size, num_frames, channels, height, width)
        # Flatten to treat frames as separate images
        batch_size, num_frames, C, H, W = pixel_values.shape
        pixel_values = pixel_values.view(batch_size * num_frames, C, H, W)
        
        # Get frame embeddings
        outputs = self.vision_model(pixel_values=pixel_values)
        pooled_output = outputs.pooler_output  # Or use last_hidden_state.mean(dim=1)
        
        # Average over frames
        pooled_output = pooled_output.view(batch_size, num_frames, -1).mean(dim=1)
        
        # Classify
        logits = self.classifier(pooled_output)
        return torch.sigmoid(logits)  # Probability for violent (1) vs non-violent (0)

# Initialize the model
fine_tuned_model = FineTunedVLM(base_model)
fine_tuned_model = fine_tuned_model.to(device)
```

### Training Setup

```python
# Hyperparameters
batch_size = 16
learning_rate = 1e-5
num_epochs = 5

# Optimizer and loss
optimizer = torch.optim.Adam(fine_tuned_model.parameters(), lr=learning_rate)
criterion = nn.BCELoss()

# DataLoaders
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
```

### Training Loop

```python
from sklearn.metrics import accuracy_score

fine_tuned_model.train()  # Set to training mode

for epoch in range(num_epochs):
    epoch_loss = 0.0
    for batch in tqdm(train_loader):
        inputs, labels = batch
        pixel_values = inputs['pixel_values'].to(device)  # Shape: (batch, num_frames, C, H, W)
        labels = labels.to(device)
        
        optimizer.zero_grad()
        outputs = fine_tuned_model(pixel_values)
        loss = criterion(outputs.squeeze(), labels)
        loss.backward()
        optimizer.step()
        
        epoch_loss += loss.item()
    
    avg_loss = epoch_loss / len(train_loader)
    print(f"Epoch {epoch+1}/{num_epochs} - Train Loss: {avg_loss:.4f}")
    
    # Validation
    fine_tuned_model.eval()
    val_preds, val_true = [], []
    with torch.no_grad():
        for batch in val_loader:
            inputs, labels = batch
            pixel_values = inputs['pixel_values'].to(device)
            outputs = fine_tuned_model(pixel_values).squeeze().cpu().numpy()
            val_preds.extend((outputs > 0.5).astype(int))
            val_true.extend(labels.cpu().numpy())
    
    val_acc = accuracy_score(val_true, val_preds)
    print(f"Validation Accuracy: {val_acc:.4f}")
    fine_tuned_model.train()  # Back to train mode
```

### Save the Model

```python
torch.save(fine_tuned_model.state_dict(), "fine_tuned_vlm.pth")
print("Model saved!")
```

### Evaluation and Inference

```python
# Load saved model for inference
fine_tuned_model.load_state_dict(torch.load("fine_tuned_vlm.pth"))
fine_tuned_model.eval()

# Example: Inference on a single video from your dataset
test_video_path = val_videos[0]  # Pick one from val set
vr = VideoReader(test_video_path, ctx=cpu(0))
frame_indices = np.linspace(0, len(vr) - 1, 8, dtype=int)
frames = vr.get_batch(frame_indices).asnumpy()

test_inputs = processor(images=list(frames), return_tensors="pt")
pixel_values = test_inputs['pixel_values'].to(device).unsqueeze(0)  # Add batch dim

with torch.no_grad():
    prediction = fine_tuned_model(pixel_values).item()

label = "violent" if prediction > 0.5 else "non-violent"
print(f"Video: {test_video_path} | Predicted: {label} (confidence: {prediction:.2f})")
```
